######1. load the iris.csv dataset.

#### 1. create a bar plot for count the frequency of the three species.

```python
### code here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("iris.csv")
df
```

```python
sns.countplot('Species',data=df)
plt.title("Iris Species Count")
plt.show()
```

##### 2. Create a pie chart to show the percentage of species.

```python
### code here
a=df.Species.value_counts()
plt.figure(figsize=(8,8))
plt.pie(x=a,labels = a.index,autopct='%0.2f%%',textprops = {'size': 'x-large'})
```

##### 3. Create the joint to all numberical variable for analyze the relationship between two variables and describe their individual distributions on the same plot.

```python
### code here
sns.jointplot(x='SepalLengthCm', y='SepalWidthCm', data=df)
sns.jointplot(x='PetalLengthCm', y='PetalWidthCm', data=df)

plt.show()
```

##### 4. Create a scatter plot and take hue = "Species"

```python
### code here
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', data=df,hue="Species")
plt.show()
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', data=df,hue="Species")
plt.show()
```

###### 5. Create a Boxplot or Whisker plot as univeriate and bivariate analysis.

```python
### code here
sns.boxplot(x='Species', y='SepalWidthCm', data=df)
```

```python
sns.boxplot(x=df['SepalLengthCm'])
```

###### 6. Create a Strip plot 

```python
### code here
plt.figure(figsize=(15,8))
sns.stripplot(x='SepalLengthCm', y='SepalWidthCm', data=df,jitter=True)

plt.show()
```

###### 7. Create a Violin Plot.

```python
### code here
sns.violinplot(x='Species', y='SepalWidthCm', data=df)
plt.show()
```

##### 8. Create a pair chart.

```python
### code here
sns.pairplot(df)
plt.show()
```

##### 9. Create a Heat Map

```python
### code here
sns.heatmap(data=df.corr(),annot=True)
plt.show()
```

###### 10. Create a distribution plot for all numerical colomns.

```python
### code here
sns.displot(df[["SepalWidthCm","SepalLengthCm",'PetalLengthCm','PetalWidthCm']],kind='kde')
plt.show()
```
